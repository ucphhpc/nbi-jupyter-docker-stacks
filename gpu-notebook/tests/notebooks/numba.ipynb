```python
import numpy as np
import numba

print(numba.__version__)
import random


@numba.jit(nopython=True)
def monte_carlo_pi(nsamples):
    acc = 0
    for i in range(nsamples):
        x = random.random()
        y = random.random()
        if (x**2 + y**2) < 1.0:
            acc += 1
    return 4 * acc / nsamples


mc_pi = monte_carlo_pi(10000)
assert mc_pi >= 0
assert mc_pi <= 4
```
